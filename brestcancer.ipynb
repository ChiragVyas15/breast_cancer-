```python
x=df.drop(columns='target',axis=1)
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
#import StandardScaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler_x_train=scaler.fit_transform(x_train)
scaler_x_test=scaler.transform(x_test)
```

```python
print(scaler_x_train)
print(scaler_x_test)
```

```python
#import gridsearchcv
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,10,100],'penalty':['l1','l2'],'solver':['liblinear']}
```

```python
#import svm
from sklearn import svm
classifier=svm.SVC(kernel='linear')
model=GridSearchCV(classifier,param_grid,scoring='accuracy',cv=5)
```

```python


from sklearn.model_selection import GridSearchCV
from sklearn import svm

param_grid = {'C': [0.1, 1, 10, 100], 'kernel': ['linear']}
classifier = svm.SVC()
model = GridSearchCV(classifier, param_grid, scoring='accuracy', cv=5)

model.fit(scaler_x_train, y_train)
```

```python
pred=model.predict(scaler_x_test)
```

```python
#plot svm curve
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test,pred))
```

```python
from sklearn.decomposition import PCA

# Reduce to 2 principal components for visualization
pca = PCA(n_components=2)
X_train_2d = pca.fit_transform(scaler_x_train)
X_test_2d = pca.transform(scaler_x_test)

# Train SVM on reduced data
svm_2d = svm.SVC(kernel='linear', C=model.best_params_['C'])
svm_2d.fit(X_train_2d, y_train)

# Plot decision boundary
plt.figure(figsize=(8,6))
x_min, x_max = X_train_2d[:, 0].min() - 1, X_train_2d[:, 0].max() + 1
y_min, y_max = X_train_2d[:, 1].min() - 1, X_train_2d[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
Z = svm_2d.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X_train_2d[:, 0], X_train_2d[:, 1], c=y_train, cmap=plt.cm.coolwarm, edgecolors='k')
plt.title('SVM Decision Boundary (PCA-reduced)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
